```python
import requests
import pandas as pd
import json
import time
from bs4 import BeautifulSoup
```

```python
import re
```

```python
from time import sleep
```

```python
pieceDetails = []

for i in range(0,10):
    sleep(1.5)
    if i == 0:
        url = 'https://www.thewatchpages.com/watches/audemars-piguet-watches/'
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        articles = soup.find_all("div", class_="px-2 pb-4 text-sm text-center skeleton-loader grow 3xl:text-base")
        for art in articles:
            details = []
            try:
                if len(art.find('div', class_="watch-price").text) > 1:
                    details.append(art.find('div', class_="mt-1 truncate title").text)
                    details.append(art.find('div', class_="mb-2 text-base truncate font-subtitle 3xl:text-lg").text)
                    priceText = art.find('div', class_="watch-price").text
                    priceValue = re.sub('\D', '', priceText)
                    #print(priceText)
                    #print(priceValue)
                    #details.append(art.find('div', class_="watch-price").text)
                    details.append(priceValue)
                    details.append(art.find('a').get("href"))
                    pieceDetails.append(details)
            except:
                details.append(art.find('div', class_="mt-1 truncate title").text)
                details.append(art.find('div', class_="mb-2 text-base truncate font-subtitle 3xl:text-lg").text)
                details.append(art.find('div', class_="italic").text)
                details.append(art.find('a').get("href"))
                pieceDetails.append(details)
    else:
        url = f'https://www.thewatchpages.com/watches/audemars-piguet-watches/page/{i}/'
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        articles = soup.find_all("div", class_="px-2 pb-4 text-sm text-center skeleton-loader grow 3xl:text-base")
        for art in articles:
            details = []
            try:
                if len(art.find('div', class_="watch-price").text) > 1:
                    details.append(art.find('div', class_="mt-1 truncate title").text)
                    details.append(art.find('div', class_="mb-2 text-base truncate font-subtitle 3xl:text-lg").text)
                    priceText = art.find('div', class_="watch-price").text
                    priceValue = re.sub('\D', '', priceText)
                    #print(priceText)
                    #print(priceValue)
                    #details.append(art.find('div', class_="watch-price").text)
                    details.append(priceValue)
                    details.append(art.find('a').get("href"))
                    pieceDetails.append(details)
            except:
                details.append(art.find('div', class_="mt-1 truncate title").text)
                details.append(art.find('div', class_="mb-2 text-base truncate font-subtitle 3xl:text-lg").text)
                details.append(art.find('div', class_="italic").text)
                details.append(art.find('a').get("href"))
                pieceDetails.append(details)
```
