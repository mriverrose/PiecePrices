```python
import requests
import pandas as pd
import json
import time
from bs4 import BeautifulSoup
```

```python
url = 'https://www.audemarspiguet.com/com/en/watch-collection.products.core-collection.json'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
json = json.loads(soup.text)
```

```python
referenceList.append([d.get('reference') for d in json['results'] if d.get('allTags')])
```

```python
referenceList
```

```python
apCurrentModels = []
for d in json['results']:
    dList = []
    dList.append(d.get('reference'))
    dList.append(d.get('collectionTitle'))
    dList.append(d.get('productTitle'))
    dList.append(d.get('caseWidth'))
    dList.append(d.get('materials'))
    apCurrentModels.append(dList)
```

```python
len(apCurrentModels)
```

```python
columns = ['reference', 'collectionTitle', 'productTitle', 'caseWidth', 'materials']
```

```python
df = pd.DataFrame(data=apCurrentModels, columns=columns)
```

```python
df
```

```python
prices = []
for l in range(0, len(apCurrentModels)):
    ref = apCurrentModels[l][0]
    priceUrl = f'https://www.audemarspiguet.com/com/en/home.price.{ref}.ch.json'
    with urllib.request.urlopen(priceUrl) as url:
        data = json.load(url)
        if len(data) == 1:
            prices.append(data['message'])
            print('price is: ', data['message'])
        elif len(data) == 2:
            prices.append(data['price']['amount'])
            print('price is: ', data['price']['amount'])
        else:
            print(f'len of data at {l} is {len(data)}')
        time.sleep(0.2)
```
